# FFT in Python

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-poster')
%matplotlib inline
```

```python
plt.style.available
```

```python
#Sampling rate
sr = 2000

#Sampling interval
ts = 1.0/sr
t = np.arange(0, 1, ts)

#Frequences
freq = 1.0
x = np.sin(2*np.pi*freq*t)

freq = 4.0
x += 5 * np.sin(2*np.pi*freq*t)

freq = 7.0 
x += np.sin(2*np.pi*freq*t)

freq = 9.0  
x += np.sin(2*np.pi*freq*t)

freq = 15.0  
x += np.sin(2*np.pi*freq*t)

plt.figure(figsize = (8, 6))
plt.plot(t, x, 'r')
plt.ylabel('Amplitude')

plt.show()
```

## FFT in Pyplot

```python
from numpy.fft import fft, ifft

X = fft(x)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.figure(figsize = (12, 6))
plt.subplot(121)

plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0, 20)

plt.subplot(122)
plt.plot(t, ifft(X), 'r')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

## FFT in Scipy

```python
from scipy.fftpack import fft, ifft

X = fft(x)

plt.figure(figsize = (12, 6))
plt.subplot(121)

plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0, 20)

plt.subplot(122)
plt.plot(t, ifft(X), 'r')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
